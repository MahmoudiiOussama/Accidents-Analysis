# ModÃ©lisation 

# 1Ã©re ModÃ¨le : Regression linÃ©aire : Variable DÃ©pendante (Y) = Nombre de victimes

#### 1Ã©re etape : Encoding les variables qualitatives

```python
from sklearn.preprocessing import LabelEncoder
object_cols = accidents_data.select_dtypes(include=['object']).columns
encoder = LabelEncoder()
accidents_data[object_cols] = accidents_data[object_cols].apply(encoder.fit_transform)
```

```python
accidents_data['Number_of_Casualties'] = accidents_data['Number_of_Casualties'].astype('int64')
```

```python
accidents_data.dtypes
```

```python
X = accidents_data.drop(["Number_of_Casualties","Date"],axis=1)
Y = accidents_data["Number_of_Casualties"]
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,train_size=0.8,random_state=200)
```

#### 2Ã©me Ã©tape : Fitting du donnÃ©es d'entraÃ®nement

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,Y_train)
```

### 4Ã©me Etape : Testing du modÃ¨le et evaluation 

```python
from sklearn.metrics import mean_squared_error 
from sklearn.metrics import accuracy_score
y_pred = model.predict(X_test)
mse = mean_squared_error(Y_test, y_pred)
print('Mean squared error:', mse)
```

#### Hyperparameters : GridSearchCV

```python
from sklearn.model_selection import  GridSearchCV
param_grid = {
    'fit_intercept': [True, False],
    'positive': [True, False],
    'copy_X': [True, False],
    'n_jobs': [ 1, 2, 4,6]}

grid_search = GridSearchCV(model, param_grid=param_grid, cv=5)

grid_search.fit(X_train,Y_train)

print('Best ParamÃ¨tres:', grid_search.best_params_)

print('Best score:', grid_search.best_score_)
```

# 2Ã©me ModÃ¨le : Classification : DecisionTreeClassifier : Variable DÃ©pendante (Y) = SÃ©veritÃ© de l'accident

### 1Ã©re etape : Split des donnÃ©es 

```python
X1 = accidents_data.drop('Accident_Severity', axis=1)
y1 = accidents_data['Accident_Severity'] 
```

```python
X1 = X1.drop("Date",axis=1)
```

#### 2Ã©me etape : Train test split

```python
from sklearn.model_selection import train_test_split
X_train1,X_test1,Y_train1,Y_test1 = train_test_split(X1,y1,train_size=0.8,random_state=42)
```

#### 3Ã©me etape: Fitting des donnÃ©es a DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier

classifier = DecisionTreeClassifier()


classifier.fit(X_train1, Y_train1)
```

#### 4Ã©me etape: Evaluation du modÃ¨le 

```python
from sklearn.metrics import accuracy_score
y_pred = classifier.predict(X_test1)
accuracy = accuracy_score(Y_test1, y_pred)
print("Accuracy: ", accuracy*100 ,"%")
```

#### Hyperparameters : RandomizedSearchCV

```python
from sklearn.model_selection import RandomizedSearchCV
param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [None, 5, 10],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': [None, 'sqrt', 'log2'],
    'min_impurity_decrease': [0.0, 0.1, 0.2]
}

random = RandomizedSearchCV(classifier, param_distributions=param_grid, n_iter=2, cv=5)

random.fit(X_train1, Y_train1)

print('Best ParamÃ¨tres:', random.best_params_)

print('Best score:', random.best_score_)
```

```python
new_classifier = random.best_estimator_

y_pred1 = new_classifier.predict(X_test1)

accuracy1 = accuracy_score(Y_test1, y_pred1)

print("Accuracy: ", accuracy1*100 ,"%")
```
